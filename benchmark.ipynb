```python
import pandas as pd
import numpy as np
```

```python
def load_data():
    train_df = pd.read_csv('Dataset/train.csv')
    test_df = pd.read_csv('Dataset/test.csv')
    val_df = pd.read_csv('Dataset/val.csv')
    train_val_df = pd.read_csv('Dataset/trainval.csv')
    return train_df, test_df, val_df, train_val_df
```

```python
if __name__ == '__main__':
   train_df, test_df, val_df, train_val_df = load_data()
   print(train_df.shape, val_df.shape,test_df.shape,train_val_df.shape)
```

```python
train_df = train_df.drop_duplicates()
```

##### now, data is clean and no null values, no duplicated values 

```python
target_train_df = train_df['Class']
train_df = train_df.drop('Class',axis=1)
target_val_df = val_df['Class']
val_df = val_df.drop('Class',axis=1)

target_train_df.sum()
```

```python
train_df['Time'].dtype
```

# Data Preprocessing

```python
train = train_df.to_numpy()
target_train = target_train_df.to_numpy()
val = val_df.to_numpy()
target_val = target_val_df.to_numpy()
test = test_df.to_numpy()
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
processor = StandardScaler()
train = processor.fit_transform(train)
val = processor.transform(val)
```

# Modeling

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(train,target_train)
pred_val = model.predict(val)
prob_val = model.predict_proba(val)[:,1]
```

# Evaluation

```python
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, classification_report

print(f'Accuracy Score is : {accuracy_score(target_val,pred_val)}')

print(f'Confustion Matrix is : \n{confusion_matrix(target_val,pred_val)}')
print(f'ROC-AUC-Score is :\n {roc_auc_score(target_val,prob_val)}')
print(f'Classification report is : \n{classification_report(target_val,pred_val)}')
```
